***ISR (indexed sparsity reduction): Miles***

```python
def isr(genes_samples, genes_filter, n_isr, cgf):
    #filter the data set to only include the "important" genes (gene_filter) if cfg is True.
    if cgf:
        genes_samples = genes_samples[finalgenes]
    
    genes_samples_isr = []
    #transpose genes_samples so it is now samples as rows and genes as columns.
    genes_samples = genes_samples.T
    for i in range(len(genes_samples)):
        sample_row = genes_samples[i]
        n_nz = np.count_nonzero(sample_row)
        top_nonzero_cgf_idx = []
        j = 0
        while len(top_nonzero_cgf_idx) < n_isr:
            if j >= len(genes_filter):
                break
            if sample_row[genes_filter[j]] != 0:
                top_nonzero_cgf_idx.append(genes_filter[j])
            j += 1
        if n_nz < n_isr:
            while len(top_nonzero_cgf_idx) < n_isr:
                top_nonzero_cgf_idx.append(0)
        genes_samples_isr.append(top_nonzero_cgf_idx)
    print(genes_samples_isr)
    return genes_samples_isr
'''
genes_samples <- matrix of the gene sample data where the genes are the rows and samples are columns. 
samples_filter <- list of samples from CFG that should be kept.
n_isr <- threshhold parameter
cgf <- if True, then isr function will return the combined result of cgf and isr. Otherwise, it only returns 
       the results of isr. 
Returns a matrix with each row being a gene and the row is the list of that gene's most important samples.
'''
result = isr(data, finalgenes, n_isr=100, cgf=False)
```
